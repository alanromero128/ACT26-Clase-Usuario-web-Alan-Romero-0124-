```python
class UsuarioWeb:
    def __init__(self, usuario, correo, contrasena):
        self.usuario = usuario
        self.correo = correo
        self.contrasena = contrasena

    def leerdatos(self):
        self.usuario = input("Ingrese el nombre de usuario: ")
        self.correo = input("Ingrese el correo electrónico: ")
        self.contrasena = input("Ingrese la contraseña: ")
        return self.usuario, self.correo, self.contrasena

    def mostrardatos(self):
        print("Usuario: ", self.usuario)
        print("Correo: ", self.correo)
        print("Contraseña: ", self.contrasena)
        return self.usuario, self.correo, self.contrasena

# Fin de la clase UsuarioWeb

print("Registro Web numero 1")
web1 = UsuarioWeb("", "", "") # Creando objeto web1
web1.leerdatos()
web1.mostrardatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*--*-*-*-*")

print("Registro Web numero 2")
web2 = UsuarioWeb("", "", "") # Creando objeto web2
web2.leerdatos()
web2.mostrardatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*--*-*-*-*")

print("Software desarrollado por Alan Romero NL = 47")
```
